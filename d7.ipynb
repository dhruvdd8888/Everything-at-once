```python
url="https://dv-h.phncdn.com/hls/videos/202208/04/413025381/,1080P_4000K,720P_4000K,480P_2000K,240P_1000K,_413025381.mp4.urlset/master.m3u8?ttl=1671540598&l=0&ipa=185.177.126.127&hash=ea1321d3142c213f91e219de4c4d1cb6"
```

```python
import requests
import m3u8


fn=url[-7:]+"dea.ts"
url.index("urlset")
prfx=url[:url.index("urlset")+7]
r=requests.get(url)

m3_Handle= m3u8.loads(r.text)

fhdlink=m3_Handle.data["playlists"][0]["uri"]
fhdlink=prfx+fhdlink

m3vid=m3u8.load(fhdlink)
io=len(m3vid.data["segments"])

with open(fn,"wb") as f:
    for seg in m3vid.data["segments"]:
        ull=prfx+seg["uri"]
        rvids=requests.get(ull)
        f.write(rvids.content)
        print(io)
        io-=1
```
